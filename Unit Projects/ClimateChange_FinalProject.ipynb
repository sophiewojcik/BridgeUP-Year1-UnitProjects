```python
#import needed libraries
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
csvFile = open("CloudFraction1_2018.csv")
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
len(data_lists)
```

```python
width = len(data_lists[0])
height = len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_values = [] # We will fill in this empty list with all the data values as we go

for r, row in enumerate(data_lists):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_values.append(float(value)) # This is where we fill in the list with all data values
        if float(value) <= 0:
            grid[r, c] = [3, 90, 252]
        elif float(value) <= 0.1:
            grid[r, c] = [36, 92, 115]
        elif float(value) <= 0.2:
            grid[r, c] = [47, 106, 130]
        elif float(value) <= 0.3:
            grid[r, c] = [60, 120, 145]
        elif float(value) <= 0.4:
            grid[r, c] = [73, 133, 158]
        elif float(value) <= 0.5:
            grid[r, c] = [101, 160, 184]
        elif float(value) <= 0.6:
            grid[r, c] = [123, 177, 199]
        elif float(value) <= 0.7:
            grid[r, c] = [150, 194, 212]
        elif float(value) <= 0.8:
            grid[r, c] = [177, 215, 230]
        elif float(value) <= 0.9:
            grid[r, c] = [194, 225, 237]
        elif float(value) <= 1.0:
            grid[r, c] = [222, 242, 250]
        else:
            grid[r, c] = [0,0,0]
```

```python
print(min(all_values))
print(max(all_values))
```

```python
plt.title("Cloud Fraction")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Cloud_Fraction_2018.png",dpi = 1000)
plt.show()
```
